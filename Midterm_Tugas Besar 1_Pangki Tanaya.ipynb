### Soal 1 (15 poin)

#### Code

```python
#import library yang akan digunakan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Soal 2 (25 poin)
Tampilkan visualisasi dari korelasi (i) antar variabel dan (ii) antara variabel dan target, lalu jelaskan hasil korelasinya.

#### Code

```python
# type your code here
corr = diabetes.corr()

ax = plt.subplots(figsize=(10,10)) 
ax = sns.heatmap(
    corr, 
    vmin=0, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    annot= True,
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```

#### Penjelasan 
Korelasi semakin kuat jika angka mendekati angka 0 (Nol)
Penentuan pengidap diabetes dikorelasikan dengan :
1. Pregnancies berkorelasi tinggi terhadap age
2. Glucose berkorelasi tinggi terhadap insulin
3. Bloodpreasure berkorelasi tinggi terhadap BMI
4. SkinThickness berkorelasi tinggi terhadap Insulin
5. DiabetesPedigreeFunction berkorelasi tinggi dengan SkinThickness

Lakukanlah _trend analysis_ nilai _daily percentage change_ dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan melalui visualisasi data.
_Trend analysis_ dilakukan dengan menggunakan hubungan berikut:
![output%2011.png](https://www.dl.dropboxusercontent.com/s/dxmrm74ksz5l9v2/trend_analysis_fig.png?dl=0)

#### Code

```python
# type your code here

#import 3 dataset
Nasdaq_df = pd.read_csv("Nasdaq.csv") 
Nikkei_df = pd.read_csv("Nikkei.csv") 
Hangseng_df = pd.read_csv("Hang_Seng.csv") 

Nasdaq_df.head()
Nikkei_df.head()
Hangseng_df.head()
```

```python
Nasdaq_df["Day_Perc_Change"] = Nasdaq_df["Adj Close"].pct_change()*100
Nasdaq_df.head()
Nikkei_df["Day_Perc_Change"] = Nikkei_df["Adj Close"].pct_change()*100
Nikkei_df.head()
Hangseng_df["Day_Perc_Change"] = Hangseng_df["Adj Close"].pct_change()*100
Hangseng_df.head()
```

```python
#Drop Value bernilai NaN
Nasdaq_df.dropna(axis = 0, inplace = True)
Nikkei_df.dropna(axis = 0, inplace = True)
Hangseng_df.dropna(axis = 0, inplace = True)
```

```python
def trend(x):
    if x > -0.5 and x <= 0.5:
        return "Slight or No change"
    elif x > 0.5 and x <= 1:
        return "Slight Positive"
    elif x > -1 and x <= -0.5:
        return "Slight Negative"
    elif x > 1 and x <= 3:
        return "Positive"
    elif x > -3 and x <= -1:
        return "Negative"
    elif x > 3 and x <= 7:
        return "Among top gainers"
    elif x > -7 and x <= -3:
        return "Among top losers"
    elif x > 7:
        return "Bull run"
    elif x <= -7:
        return "Bear drop"
    
Nasdaq_df["Trend"]=Nasdaq_df["Day_Perc_Change"].apply(lambda x:trend(x))
Nasdaq_df.head()
Nikkei_df["Trend"]=Nikkei_df["Day_Perc_Change"].apply(lambda x:trend(x))
Nikkei_df.head()
Hangseng_df["Trend"]=Hangseng_df["Day_Perc_Change"].apply(lambda x:trend(x))
Hangseng_df.head()
```

```python
Nasdaq_df["Day_Perc_Change"] = Nasdaq_df["Adj Close"].pct_change()*100
Nasdaq_df.head()

Nasdaq_pie_data = Nasdaq_df.groupby('Trend')
pie_label = sorted([i for i in Nasdaq_df.loc[:, 'Trend'].unique()])
plt.pie(Nasdaq_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

#### Penjelasan
Berdasarkan grafik nasdaq diatas dapat disimpulkan bahwa perubahan tiap hari masuk kategori : 
1. Slight or no change dengan presentase 35%
2. Positif dengan presentase 21.2%
3. Slight positif dengan presentase 16%
4. Negatif dengan presentase 12%
5. Slight negatif dengan presentase 8.2%
6. Among top losers dengan presentase 4.4%
7. Among top gainers dengan presentase 2%
8. Bull run dan Bear drop dengan presentase 0.6%

Harga saham cenderung tidak berubah namun nilai positif cukup besar sehingga direkomendasikan untuk jangka panjang

```python
Nikkei_df["Day_Perc_Change"] = Nikkei_df["Adj Close"].pct_change()*100
Nikkei_df.head()

Nikkei_pie_data = Nikkei_df.groupby('Trend')
pie_label = sorted([i for i in Nikkei_df.loc[:, 'Trend'].unique()])
plt.pie(Nikkei_pie_data['Trend'].count(), labels = pie_label, 
        autopct = '%1.1f%%', radius = 2)

plt.show()
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
# merge Adj. close values
combined_df = pd.concat([Nasdaq_df["Adj Close"], Nikkei_df["Adj Close"], Hangseng_df["Adj Close"]], axis=1)
combined_df.columns = ["NASDAQ.NS", "NIKKEI.NS", "HANGSENG.NS"]
```

```python
# drop null values
combined_df.dropna(inplace = True, axis = 0)

# display first few rows
combined_df.head()
```

```python
# store daily returns of all above stocks in a new dataframe 
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)

# plotting pairplot  
import seaborn as sns
sns.set(style = "ticks", font_scale = 1.25)
sns.pairplot(pct_chg_df)
```

```python
Nasdaq_vol = pct_chg_df["NASDAQ.NS"].rolling(7).std()*np.sqrt(7)
Nasdaq_vol.plot(figsize = (15, 7))
```

```python
Nikkei_vol = pct_chg_df["NIKKEI.NS"].rolling(7).std()*np.sqrt(7)
Nikkei_vol.plot(figsize = (15, 7))
```

```python
Hangseng_vol = pct_chg_df["HANGSENG.NS"].rolling(7).std()*np.sqrt(7)
Hangseng_vol.plot(figsize = (15, 7))
```

```python
volatility = pct_chg_df[['NASDAQ.NS', 'NIKKEI.NS', 'HANGSENG.NS']].rolling(7).std()*np.sqrt(7)
volatility.plot(figsize = (15, 7))
```

#### Penjelasan 
Berdasarkan grafik diatas peubahan umumnya tidak terlalu signifikan namun terjadi kenaikan tertinggi pada Nasdaq dan penurunan terendah pada saham Nikkei
